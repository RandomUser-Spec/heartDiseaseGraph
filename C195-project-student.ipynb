# Task 1 - Find the diabetic and hight blood pressure patients across all age groups, and conclude the risk heart stroke is higher in which age group

```python
#Import libraries
import pandas as pd
import matplotlib.pyplot as plt 

#read the csv
df = pd.read_csv('heart_disease.csv')
df
```

```python
#Filter by diabetes(condition will be who has diabetes) and create new dataframe
diabetespatient = df.loc[df['diabetes'] == 1]
diabetespatient
```

```python
#On this new data frame perform group operation as per age and create new dataframe 
group_operation_age = diabetespatient.groupby('age')['diabetes'].count().reset_index()
group_operation_age
```

```python
#Filter by high_blood_pressure(condition will be who has high_blood_pressure) and create new dataframe
high_blood_pressure = df.loc[df['high_blood_pressure'] == 1]
high_blood_pressure
```

```python
#On this new data frame perform group operation as per age and create new dataframe 
high_blood_pressure_per_age = high_blood_pressure.groupby('age')['high_blood_pressure'].count().reset_index()
high_blood_pressure_per_age
```

```python
#plot the scatter graph to show which age group is more prone to diabetes
plt.figure(figsize = (19,8))
diabetes = group_operation_age['diabetes']
age = group_operation_age['age']

plt.scatter(age, diabetes, label = 'Diabetes Patient', marker = '^', c = "red", s = 250)

age2 = high_blood_pressure_per_age['age']
blood_pressure = high_blood_pressure_per_age['high_blood_pressure']

plt.scatter(age2, blood_pressure, label = "Blood Pressure", marker = 's', c = "yellow", s = 250)

plt.legend()
```

# Task 2 - Find as per gender who has not normal platelets level in blood

```python
#Filter by platelets(condition lesser then 150000 OR greater then 450000) and create new dataframe
filtered_platelets = df.loc[(df['platelets'] <= 150000.00) | (df['platelets'] >= 450000.00)]
filtered_platelets
```

```python
#On this new dataframe perform group operation as per gender and create new dataframe 
group_operation_per_gender = filtered_platelets.groupby('gender')['platelets'].count().reset_index()
group_operation_per_gender


```

```python
#Plot a pie chart as per the gender to show the percentage of male and female who has not normal platelets
value = group_operation_per_gender['platelets']
label = group_operation_per_gender['gender']

plt.pie(value, labels = label, autopct = '%0.1f%%', radius= 2)
plt.show()
```
